### Importando dataset

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 
import os
```

```python
file_path = %pwd
file_name = 'trump-tweets.csv'

df = pd.read_csv(file_name, sep=',', header=0)
```

`Coluna "Client Simplified" dropada do dataset pois a mesma não tem muita importância neste contexto`

### Tratamento de dados

*`É retirada qualquer caracter especial e palavras contidas na lista de stopwords`

```python
import re
```

```python
stopWords = open('stopwords.txt','r').read().split('\n')
```

```python
tweets = []

for rows in df['Tweet']:
    
    tweet = [word.upper() for word in rows.split() if word.lower() not in stopWords]
    tweet = ' '.join(tweet)
    tweet = re.sub('[^a-zA-Z0-9 ]+','',tweet)
    tweets.append(tweet)
```

```python
def show(wordCloud, s1, s2):
    plt.figure(figsize=(s1,s2))
    plt.axis('off')
    plt.imshow(wordCloud, interpolation='bilinear')
```

##  Este é um exemplo default de como é gerado uma wordcloud

<div class="alert alert-block alert-info">
<b>OBS:</b> Você pode usar fontes personalizadas nas imagens, basta baixar seu arquivo .ttf, coloca-la na pasta "fonts" e renomear o arquivo font abaixo
</div>


```python
font = 'Calibri.ttf'
fontpath = os.path.join('fonts',font)
```

```python
import wordcloud

wc = wordcloud.WordCloud(font_path=fontpath,
    width=1600, height=800,
    background_color='black',
    max_words=1000,
    max_font_size=1000
).generate(' '.join(tweets))
```

```python
show(wc,15,10)
```

<div class="alert alert-block alert-info">
<b>OBS:</b> Existem maneiras de usar cores customizadas, no exemplo abaixo é criada uma função que retorna cores brancas e cinzas para a imagem, onde é feito apenas seu coloramento

Você pode usar o site da W3School (https://www.w3schools.com/colors/colors_hsl.asp) para escolher as cores que deseja.

No metodo é gerador um range de valores aleatórios no terceiro parametro que você pode escolher
</div>

```python
import numpy as np
```

```python
def grey_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(10,0%%, %d%%)" % np.random.randint(30,100))
```

```python
wc.recolor(color_func=grey_color_func)
```

```python
show(wc,15,10)
```

<div class="alert alert-block alert-info">
<b>OBS:</b> Existem também formas de fazer as imagens com formatos diferentes
    
Neste exemplo eu usei a bandeira dos Estados Unidos, porém você pode baixar sua propria imagem e alterar o valor da variavel com o nome da imagem que você baixou
</div>

```python
from PIL import Image
from wordcloud import ImageColorGenerator
```

```python
img_usa = 'usa-flag.png'
usa_mask = np.array(Image.open(os.path.join(file_path, img_usa)))
```

```python
wc = wordcloud.WordCloud(font_path=fontpath,
    mask=usa_mask,
    background_color='black',
    width=1600, height=800,
    max_words=1000,
    max_font_size=1000
).generate(' '.join(tweets))
```

```python
image_colors = ImageColorGenerator(usa_mask)
wc.recolor(color_func=image_colors)
```

```python
show(wc,15,10)
```

### Exemplo final utilizando todos os metodos acima

```python
import multidict

img_thumb = 'thumbs-up.jpg'
thumb_mask = np.array(Image.open(os.path.join(file_path, img_thumb)))
```

```python
words = multidict.MultiDict()
for i in range(200):
    words.add('Obrigado', np.random.randint(1, 10))
```

```python
font = 'MonotypeCorsiva.ttf'
fontpath = os.path.join('fonts',font)
```

```python
wc = wordcloud.WordCloud(font_path=fontpath,
    mask=thumb_mask,
    background_color='white',
    width=1600, height=800,
    max_font_size=1000).generate_from_frequencies(words)
```

```python
def green_color_func(word, font_size, position,orientation,random_state=None, **kwargs):
    return("hsl(120,100%%, %d%%)" % np.random.randint(20,40))
```

```python
image_colors = ImageColorGenerator(thumb_mask)
wc.recolor(color_func=green_color_func)
```

```python
show(wc,10,6)
```
